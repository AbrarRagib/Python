````python
import os
import json
````

````python
path = "../data/autolabeled"
````

Complete Schema:
```json
[
  {
    "bbox": [x1, y1, x2, y2],           // Required: 4 integers, x1<x2, y1<y2
    "category": "Text",                  // Required: One of 11 valid categories
    "text": "content",                   // Required (except for 'Picture')
    "font_size": 12,                     // Required: Integer 8-26
    "font_style": "Normal",                // Required: String (Normal, Bold, Italic, Underline)
    "alignment": "left",                 // Required: (bbox-level alignment) "left"|"center"|"right"|"justify"
    "page_alignment": "left"             // Required: (page-level alignment of that bbox) "left"|"center"|"right"
  }
]
```

````python
def process_annotation(annot_obj):

    updated_obj = {}
    updated_obj["bbox"] = annot_obj["bbox"]
    updated_obj["category"] = annot_obj["category"] 
    if updated_obj["category"]!="Picture":
        updated_obj["text"] = annot_obj["text"]
        updated_obj["font_size"] = ""
        updated_obj["font_style"] = ""
    updated_obj["alignment"] = annot_obj["alignment"]
    updated_obj["page_alignment"] = ""

    return updated_obj
````

````python
for root, _, fnames in os.walk(path):
    for fname in fnames:
        if fname.endswith(".json"):
            
            print(f"Processing: {fname}")

            with open(os.path.join(root, fname), "r") as fr:
                data = json.load(fr)
            
            data_updated = [process_annotation(entry) for entry in data]

            with open(os.path.join(root, fname), "w") as fw:
                json.dump(data_updated, fw, ensure_ascii=False)
````
